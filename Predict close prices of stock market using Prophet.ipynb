```python
import pandas as pd
import numpy as np
from datetime import datetime

import matplotlib.pyplot as plt
import seaborn as sns

#Create the grid (background) in plot charts
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline

start_date = "2018-12-31"
#end_date = "2023-01-18"


# Collect the current system date
now = datetime.now()

# Get the current date in the format YYYY-MM-DD
end_date = now.strftime("%Y-%m-%d")


cbaax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/CBA.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
cbaax = cbaax.loc[start_date:end_date]

wbcax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/WBC.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
wbcax = wbcax.loc[start_date:end_date]

anzax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/ANZ.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
anzax = anzax.loc[start_date:end_date]

wesax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/WES.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
wesax = wesax.loc[start_date:end_date]


```

```python
# Import the necessary libraries
# Prophet is a package to predict stock future trends
from fbprophet import Prophet

#to stop uncessary warning use two below coding lines
import logging
logging.getLogger('fbprophet').setLevel(logging.WARNING) 

# Normally before plotting the stock data using Prophet, we need to convert the date column to datetime format using:
#cbaax['Date'] = pd.to_datetime(cbaax['Date'])

# Then set the date column as the index
#cbaax.set_index('Date', inplace=True)

#It's important to make sure that the ds column is in datetime format and that the y column is numerical. You can use the to_datetime() and to_numeric() functions from the pandas library to change the datatype of the columns:

# However, the data I have here already in datetime format so I don't need to go through the step above

# Create a new DataFrame with the columns 'ds' and 'y'.
df_prophet = pd.DataFrame({'ds': cbaax.index, 'y' : cbaax['Close']})

df_prophet = df_prophet.rename(columns = {'Date' : 'ds', 'ClosePrice' : 'y'})

# Initialize the Prophet model
model = Prophet()

# Fit the model to the data
model.fit(df_prophet)

# Create a future DataFrame with the specified number of days
future = model.make_future_dataframe(periods=31)

# Make predictions
forecast = model.predict(future)

# Plot the predictions with model.plot(forecast);
# The ; to stop the charts appear twice (sometimes this happens)

model.plot(forecast)

# Make the blue line thiner
# Get the line object for the forecast line
line = plt.gca().get_lines()[1]

# Set the line thickness
line.set_linewidth(2)

#change chart label and add title
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.title('Close Price Prediction for CBAAX')
```

```python
# Predict stock prices in the next 3 months

#Title of the figure
#fig.suptitle('Predicting stock prices in the next three months', fontsize=36)

#to stop uncessary warning use two below coding lines
import logging
logging.getLogger('fbprophet').setLevel(logging.WARNING) 

# Plot the data for all subplots
df_prophet_cba = pd.DataFrame({'ds': cbaax.index, 'y' : cbaax['Close']})
df_prophet_wbc = pd.DataFrame({'ds': wbcax.index, 'y' : wbcax['Close']})
df_prophet_anz = pd.DataFrame({'ds': anzax.index, 'y' : anzax['Close']})
df_prophet_wes = pd.DataFrame({'ds': wesax.index, 'y' : wesax['Close']})

# Initialize the Prophet model
model_cba = Prophet()
model_wbc = Prophet()
model_anz = Prophet()
model_wes = Prophet()

# Fit the model to the data
model_cba.fit(df_prophet_cba)
model_wbc.fit(df_prophet_wbc)
model_anz.fit(df_prophet_anz)
model_wes.fit(df_prophet_wes)

# Create a future dataframe with the specified number of days
future_cba = model_cba.make_future_dataframe(periods=365)
future_wbc = model_wbc.make_future_dataframe(periods=365)
future_anz = model_anz.make_future_dataframe(periods=365)
future_wes = model_wes.make_future_dataframe(periods=365)

# Make predictions
forecast_cba = model_cba.predict(future_cba)
forecast_wbc = model_wbc.predict(future_wbc)
forecast_anz = model_anz.predict(future_anz)
forecast_wes = model_wes.predict(future_wes)

# Get the range of the forecast predictions
start_date = forecast_cba['ds'].iloc[-365]
end_date = forecast_cba['ds'].iloc[-1]

start_date = forecast_wbc['ds'].iloc[-365]
end_date = forecast_wbc['ds'].iloc[-1]

start_date = forecast_anz['ds'].iloc[-365]
end_date = forecast_anz['ds'].iloc[-1]

start_date = forecast_wes['ds'].iloc[-365]
end_date = forecast_wes['ds'].iloc[-1]

# Plot only the forecast predictions
plt.plot(forecast_cba['ds'], forecast_cba['yhat'],label = 'CBAAX', linewidth = 1)
plt.plot(forecast_wbc['ds'], forecast_wbc['yhat'],label = 'WBCAX', linewidth = 1)
plt.plot(forecast_anz['ds'], forecast_anz['yhat'],label = 'ANZAX', linewidth = 1)
plt.plot(forecast_wes['ds'], forecast_wes['yhat'],label = 'WESAX', linewidth = 1)


plt.xlim([start_date, end_date])
plt.legend()
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.title('Predict close prices in the next year')
ax = plt.xticks(rotation=45)

# Adjust the spacing between the subplots

#plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.85, wspace=0.2, hspace=0.25)

# Show the plot

plt.show()
```
